```python
import numpy as np
```

```python
a=np.arange(1,26).reshape(5,5)
print(a)
```

```python
a[:2,:3] #first 2 rows and First 3 columns
```

```python
a[:,3:]  #all rows and last 2 columns
```

```python
a[1:,:2]  #all rows except row no 0, first 2 columns
```

```python
a[2:,2:] #all rows and columns excepting first 2 rows and columns
```

```python
a[1:4,1:4] #all rows from row no 1 to 4 except 4, columns from row no 1 to 4 except 4
```
